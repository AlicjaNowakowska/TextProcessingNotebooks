For a given bitstring b list all bitstrings b’, such that the Hamming distance between b and b’ is equal 1.
b' should differ from b only at one position, so given:
cat 
b' will be:
dowolna_litera a t
c dowolna_litera t
c a dowolna_litera 

```python
%pprint
```

```python
def zad21(b):
    letters=["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
    b_prim=[]
    for i in range(0,len(b)):
        for j in letters: 
            b_prim.append(b[0:i]+j+b[i+1:])
    return b_prim
```

Construct a function that returns a Jaccard similarity for two sets. Beware that this

Jaccard similarity intersection(A,B)/sum(A,B)

```python
def zad22(A,B):
    A=set(A)
    B=set(B)
    try:
        jaccard=len(A.intersection(B))/len(A.union(B))
        print("A.intersection(B)",A.intersection(B))
        print("A.union(B)",A.union(B))
        return jaccard
    except:
        print("not correct arguments")
```

### zad 23
Construct a function that computes Jaccard similarity for two strings treated as bags
of words.

```python
def zad23(A,B):
    A=list(A)
    B=list(B)
    sum_A_B=A+B
    #print("bag sum A and B",sum_A_B)
    #inter_A_B=set(A).intersection(set(B))
    inter_A_B=list(filter(lambda x:x in A, B))
    #print("bag intersection A and B", inter_A_B)
    jaccard=len(inter_A_B)/len(sum_A_B)
    return jaccard
```

```python
zad23("ryybka","myyszka")
```

### zad 26.
Let us consider a metric space (S; d), where S is the set of words from
text1 and d is the Hamming distance. Find diameter of (S; d).

A metric space M is called bounded if there exists some number r, such that d(x,y) ≤ r for all x and y in M. The smallest possible such r is called the diameter of M.

```python
import nltk
nltk.download()
from nltk.book import *
set_text1=set(text1) #all words
```

```python
#szukam sobie najmniejszego ograniczenia z góry
diameter=1
for i in set_text1:
    for j in set_text1:
        if i !=j and len(i)==len(j): #must have equal length
            max_hamming=sum(ex!=ey for ex,ey in zip(i,j)) #calculate distance
            if max_hamming>diameter: #is it minimal distance?
                diameter=max_hamming
print(diameter)
```

### zad 28
Draw a graph with nodes labeled by words in text2 that appear at
least l times. Add edges conecting pairs of words with edit distance smaller than s.
Try to minimize l, maximize s and keep the quality of your visualization (networkx
may be insufficient).

```python
import networkx as nx
def zad28(l,s):
    set_text2=set(text2) #all words
    G = nx.Graph()
    nodes=[i for i in set_text2 if text2.count(i)>l ]
    G.add_nodes_from(nodes)
    edges=[(i,j) for i in nodes for j in nodes if i!=j and nltk.edit_distance(i,j)<s]
    G.add_edges_from(edges)
    return(G)
    #nx.draw(G)
```

```python
G=zad28(25,3)
```

```python
plt.figure(figsize=(30,20)) 
#import matplotlib.pyplot as plt
d = dict(G.degree)
#pos = nx.spring_layout(G, iterations=200)
pos = nx.spring_layout(G)#, iterations=200)
nx.draw(G,pos,node_size=10,alpha=0.5,node_color=range(len(G.nodes)),cmap=plt.cm.Blues)
#, node_size=1, node_color="red", alpha=0.5, linewidths=40)
nx.draw(G,pos,label=True,nodelist=d.keys(), node_size=[v * 100 for v in d.values()],alpha=0.7,
        node_color=range(len(G.nodes)),cmap=plt.cm.Blues)
plt.title("Dark and big nodes are those with many connections, light and small with few edges")
```

### zad 29
For two words v and w, let relative edit distance be the Levensthein
distance between v and w divided by the sum of lengths v and w. Find two different
words in text2 with minimal relative edit distance.

https://www.occasionalenthusiast.com/efficient-recursive-levenshtein-edit-distance-algorithm/
    using recursive algorithm for the calculation of the levensthein distance/

```python
def leve(i,j,v,w):# levensthein distance for sequences of length i (len(v)) and j len(w)
    
    if min(i,j)==0:
        return max(i,j)
    else:
        if v[i-1]==w[j-1]:
            k=0
        else:
            k=1
        
        return(min(leve(i-1,j,v,w)+1,leve(i,j-1,v,w)+1,leve(i-1,j-1,v,w)+k))
    
def relative_leve(v,w):
    distance=leve(len(v),len(w),v,w)
    return distance/(len(v)+len(w))
```

### Zad 30
For a given bitstring b and a natural number n list all bitstrings b’, such that the
Hamming distance between b and b’ is equal n.

bitstring=0111010101

```python
def zad30(word,n):
    l=it.product(*['01']*len(word))
    l=list(l) # it gives me string in the form s= ("0","1", "0"), ''.join(s)  joins s to "010"
    #now all those that differ at n sites
    l=[''.join(i) for i in l if sum(x!=y for x,y in zip(word,''.join(i)))==n]
    return l
```

```python
zad30("0101",2)
```

```python
zad30("0101",3)
```

### zad 31.
Construct a function that for a given string and a natural number k returns a set
of all its k-shingles.

```python
def zad31(string,k):
    shingles=[]
    for i in range(len(string)-k):
        shingles.append(string[i:i+k])
    return set(shingles)
```

```python
zad31("abracadabra",3)
```
