### Zad 12. 
Implement a Cesar cipher using with a given shift parameter d in Python. Use functions translate() and maketrans()

```python
def cesar_coding(text, d):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    translate_table = str.maketrans(alphabet, alphabet[d:] + alphabet[:d]) #literki - > na literki
    return text.translate(translate_table)
```

```python
print("text to encrypt: ",'kodowanie szyfru')
print("encrypted text with shift parameter 2: ",cesar_coding("kodowanie szyfru",2))
```

### a) Remove all the punctuation

```python
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(file) #removes punctuation, creates list of words
removed_punctuation=" "
for i in range(500,1000):
    removed_punctuation+=words[i]+" "
print("Sample text of Jane Asuten with removed punctuation:".upper())
print(removed_punctuation)
```

### b)  Remove all stop words

```python
stopWords=sorted(nltk.corpus.stopwords.words('english'))
stopWords[0:20]
```

```python
removed_stop_words=[i for i in words if i.lower() not in stopWords]
```

```python
removed=" "
for i in range(1000,1500):
    removed+=removed_stop_words[i]+" "
print("Sample text of Jane Asuten with removed punctuation and stop words:".upper())
print(removed)
```

### c) Remove infected words to their word stem (stemming words).

```python
from nltk.stem import PorterStemmer 
ps = PorterStemmer() 
removed_stemming_words = [ps.stem(w) for w in words]
```

```python
removed=" "
for i in range(1000,1500):
    removed+=removed_stemming_words[i]+" "
print("Sample text of Jane Asuten with removed punctuation and stemming words:".upper())
print(removed)
```

### d ) List all the words that appear at least 100 times in the text. Than sort them in alphabetical order.

```python
counters={}
appearing_words=set(words)
for i in appearing_words:
    counters[i]=words.count(i)
```

### In alphabetical order

```python
sorted(counters.items())
```

### Appear at least 100 times

```python
at_least100=[i for i in counters if counters[i]>=100]
at_least100
```

### Zad 14
Construct a dictionary that assigns each user (represented as a string) his address
(string). Sort and list this dictionary by the addresses. Use lambda function.

```python
users=["Monika","Julia","Kasia","Marek","Kocur","Ala"]
adress=["home","kitchen","room","swimming pool","bed","tropical island"]
dictionary={}
for i in range(0,len(users)):
    dictionary[users[i]]=adress[i]
```
